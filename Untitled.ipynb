```python
#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv file
df = pd.read_csv("flights.csv")
```

```python
df = df.dropna()
```

```python
plt.hist(df["arr_delay"])
```

```python
sns.regplot(x='arr_delay',y='dep_delay',data=df)
```

```python
df["carrier1"] = pd.Categorical(df["carrier"], df.carrier.unique())
df["origin1"] = pd.Categorical(df["origin"], df.origin.unique())
flight = df.pivot('carrier1', 'origin1', 'arr_delay')
ax = sns.heatmap(flight)
```
